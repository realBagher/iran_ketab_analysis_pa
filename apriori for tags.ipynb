```python
import pandas as pd
from apyori import apriori
```

```python
tags=pd.read_csv('tags all.csv')
tags.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
df=tags.groupby(['book_id'])['name'].agg(lambda x : list(x))
```

```python
transactions = list(df)
results = list(apriori(transactions, min_support=0.008))

rules_df = pd.DataFrame(columns=('Antecedent','Consequent','Confidence'))

Confidence = []
Antecedent = []
Support = []
Consequent=[]

for RelationRecord in results:
    for ordered_stat in RelationRecord.ordered_statistics:
        Support.append(RelationRecord.support)
        Antecedent.append(ordered_stat.items_base)
        Consequent.append(ordered_stat.items_add)
        Confidence.append(ordered_stat.confidence)

# df['Items'] = list(map(set, Items))                                   
rules_df['Antecedent'] = list(map(set, Antecedent))
rules_df['Consequent'] = list(map(set, Consequent))
rules_df['Support'] = Support
rules_df['Confidence'] = Confidence
```

```python
rules_df['len_act']=rules_df.Antecedent.apply(lambda x : len(x))
```

```python
rules_df['act']=rules_df.Antecedent.apply(lambda x : ','.join(p for p in x))
```

```python
rules_df['len_cqt']=rules_df.Antecedent.apply(lambda x : len(x))
```

```python
rules_df['cqt']=rules_df.Antecedent.apply(lambda x : ','.join(p for p in x))
```
